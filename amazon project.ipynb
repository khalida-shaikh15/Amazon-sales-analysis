```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df= pd.read_csv('Amazon Sales data.csv')
```

```python
df
```

```python
df.info()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Extract year, month, and year-month
df['year'] = df['Order Date'].dt.year
df['month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day
df['year_month'] =  df['Order Date'].dt.strftime('%y-%m')
```

```python
df
```

```python
df.duplicated().sum()
```

```python
df.groupby('month')['Units Sold'].sum().reset_index()
```

```python
monthly_sales = df.groupby('month')['Units Sold'].sum().reset_index()
yearly_sales = df.groupby('year')['Units Sold'].sum().reset_index()
yearly_monthly_sales = df.groupby('year_month')['Units Sold'].sum().reset_index()
```

```python

sns.lineplot(data=monthly_sales, x='month', y='Units Sold', marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
```

```python

sns.lineplot(data=yearly_sales, x='year', y='Units Sold', marker='o')
plt.title('yearly Sales Trend')
plt.xlabel('year')
plt.ylabel('Total Sales')
```

```python
plt.figure(figsize=(25,10))
plt.plot(yearly_monthly_sales['year_month'],yearly_monthly_sales['Units Sold'])
plt.title('Monthly yearly Sales Trend')
plt.xlabel('Month and year')
plt.ylabel('Total Sales')
```

```python
 plt.subplot(1, 2, 1)

sns.lineplot(data=df, x='year', y='Total Profit',marker='o')
plt.title('yearly profit Trend')
plt.xlabel('year')
plt.ylabel('Total Sales')

plt.subplot(1,2 , 2)
sns.lineplot(data=df, x='month', y='Total Profit',marker='o')
plt.title('monthly profit Trend')
plt.xlabel('year')
plt.ylabel('Total Sales')

plt.tight_layout(pad=1.0)
```

```python
tempdf=df[["Units Sold","Unit Price","Unit Cost","Total Revenue","Total Cost","Total Profit"]]
correlation_matrix = tempdf.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt2=df.groupby("year")['Total Profit'].sum()
plt.subplot(1, 2, 1)
plt.plot(plt2)
plt.title('yearly total Profit')
plt.xlabel('year')
plt.ylabel('Total profit')

plt1=df.groupby("year")['Order ID'].nunique()
plt.subplot(1, 2, 2)
plt.plot(plt1)
plt.title('yearly total sales')
plt.xlabel('year')
plt.ylabel('Total Sales')

plt.tight_layout(pad=1.0)
```
